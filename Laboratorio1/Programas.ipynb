```python
#Torres de Hanoi
"""Input------- Int. Número de discos que contiene el juego
   Output------- Pasos a hacer para resolver el problema"""

def torres_Hanoi(discos, inicial, auxiliar, final):

  if(discos <= 1):
    print('Disco', str(discos), "de torre ", inicial, "a torre ", final)

  else:
    #Mueve los primeros n-1 discos de la torre inicial a la torre auxiliar
    torres_Hanoi(discos-1, inicial, final, auxiliar)
    #Mueve el disco más grande de la torre inicial a la final
    print('Disco', str(discos), "de", inicial, "a", final)
    #Mueve los discos que estaban en la torre auxiliar a la final
    torres_Hanoi(discos-1, auxiliar, inicial, final)

torres_Hanoi(3, "inicial", "auxiliar", "final")
```

```python
#Algoritmo recursivo que resuelve los triominos
"""Input: Integer. Potencia a la que se va a elevar el número 2 para obtener un tablero 2^n x 2^n
   Output: Pasos a hacer para embaldosar el tablero con triominós"""
def triominos(n):

  #Si el tablero es de 2x2 entonces unicamente coloca la pieza y listo
  if n ==  1:
    print("Paso base")
  else:
    print(f"Resolviendo para n= {n} ")
    #Divide el tablero en 4 de igual longuitud
    print("Dividiendo...")
    #Pone la pieza en el centro del tablero original
    print("Poniendo pieza central")
    #Llama recursivamente 4 veces a triominos(n-1) para resolver los otros tableros
    print(f"Resolviendo la primera división")
    triominos(n-1)
    print(f"Resolviendo la segunda división")
    triominos(n-1)
    print(f"Resolviendo la tercera división")
    triominos(n-1)
    print(f"Resolviendo la cuarta división")
    triominos(n-1)
    print("Fin :)")
```
